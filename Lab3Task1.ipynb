```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('D://THINK/Downloads/Salary_dataset.csv')
print("Dataset:\n", data)
```

```python
X = data[['YearsExperience']]
Y = data['Salary']
```

```python
model = LinearRegression()
model.fit(X,Y)
```

```python
predicted_score = model.predict([[3]])
print("Predicted score for 3 years of experience:", predicted_score[0])
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
Y_pred = model.predict(X)
print("Mean Squared Error:", mean_squared_error(Y, Y_pred))
```

```python
plt.scatter(X, Y, color='green', label='Actual Data')
plt.plot(X, model.predict(X), color='blue', label='Regression Line')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Years Experience vs Salary')
plt.legend()
plt.show()
```
